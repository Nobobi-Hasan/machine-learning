```python
# X = df.iloc[:,:-1]
X = df[['height', 'weight', 'shoe size']]
X.head()
```

```python
# X = df_train[['feature1', 'feature2']]  # Training data with named columns
# model.fit(X, y)

model = tree.DecisionTreeClassifier()
model.fit(X, y)
pred = model.predict([[171, 65, 41]])
pred
```

## K-NN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_C_model =  KNeighborsClassifier(n_neighbors = 3)
```

```python
KNN_C_model.fit(X, y)
```

```python
# pred_KNNC = KNN_C_model.predict([[171, 65, 41]])
pred_KNNC = KNN_C_model.predict(X)
pred_KNNC
```

```python
KNN_C_model.score(X,y)
```

## K-NN Regressor (Height Prediction)

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encoding
label_encoder = LabelEncoder()
df['gender'] = label_encoder.fit_transform(df['gender'])

df.head()
```

```python
X_R = df.drop(['height'], axis = 1)
y_r = df['height']
```

```python
KNN_R_model =  KNeighborsRegressor(n_neighbors = 3)
KNN_R_model.fit(X_R, y_r)
```

```python
KNN_R_model.predict(X_R)
```

```python
KNN_R_model.score(X_R, y_r)
```
